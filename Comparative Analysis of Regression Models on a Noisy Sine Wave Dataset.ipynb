# Comparative Analysis of Regression Models on a Noisy Sine Wave Dataset

## 1. Import Required Libraries`

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

## 2. Generate Synthetic Sine Wave Dataset

```python
# Set seed for reproducibility
np.random.seed(42)

# Generate data
X = np.linspace(0, 2 * np.pi, 120).reshape(-1, 1)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, 120)
```

##  3. Visualize the Dataset

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='deepskyblue', label='Noisy Data', s=25)
plt.plot(X, np.sin(X), color='black', linestyle='--', label='True Sine', linewidth=2)
plt.title('Noisy Sine Wave Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('sine_data.png')
plt.show()
plt.close()
```

## 4. Initialize and Train Regression Models

```python
# Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(X, y)

# Polynomial Regression (degree 3)
polyreg = make_pipeline(PolynomialFeatures(degree=3), LinearRegression())
polyreg.fit(X, y)

# Random Forest Regressor
rf_reg = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
rf_reg.fit(X, y)

# MLP Regressor
mlp_reg = MLPRegressor(hidden_layer_sizes=(50, 50), activation='relu', solver='adam',
                       max_iter=2000, random_state=42)
mlp_reg.fit(X, y)
```

## 5. Make Predictions

```python
y_pred_linear = linear_reg.predict(X)
y_pred_poly = polyreg.predict(X)
y_pred_rf = rf_reg.predict(X)
y_pred_mlp = mlp_reg.predict(X)
```

##  6. Evaluate Model Performance & Visualize Model Predictions

# Linear Regression

```python
mse = mean_squared_error(y, y_pred_linear)
r2 = r2_score(y, y_pred_linear)
print('Linear Regression:')
print(f'  MSE: {mse:.4f}')
print(f'  R² Score: {r2:.4f}')

plt.figure(figsize=(6, 4))
plt.scatter(X, y, color='deepskyblue', label='Actual Data', s=25)
plt.plot(X, y_pred_linear, color='darkred', label='Linear Predictions', linewidth=2)
plt.plot(X, np.sin(X), color='black', linestyle='--', label='True Sine', linewidth=1)
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('linear_regression.png')
plt.show()
```

## Polynomial Regression

```python
mse = mean_squared_error(y, y_pred_poly)
r2 = r2_score(y, y_pred_poly)
print('Polynomial Regression:')
print(f'  MSE: {mse:.4f}')
print(f'  R² Score: {r2:.4f}')

plt.figure(figsize=(6, 4))
plt.scatter(X, y, color='deepskyblue', label='Actual Data', s=25)
plt.plot(X, y_pred_poly, color='darkred', label='Polynomial Predictions', linewidth=2)
plt.plot(X, np.sin(X), color='black', linestyle='--', label='True Sine', linewidth=1)
plt.title('Polynomial Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('polynomial_regression.png')
plt.show()
```

## Random Forest Regression

```python
mse = mean_squared_error(y, y_pred_rf)
r2 = r2_score(y, y_pred_rf)
print('Random Forest Regression:')
print(f'  MSE: {mse:.4f}')
print(f'  R² Score: {r2:.4f}')

plt.figure(figsize=(6, 4))
plt.scatter(X, y, color='deepskyblue', label='Actual Data', s=25)
plt.plot(X, y_pred_rf, color='darkred', label='RF Predictions', linewidth=2)
plt.plot(X, np.sin(X), color='black', linestyle='--', label='True Sine', linewidth=1)
plt.title('Random Forest Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('random_forest_regression.png')
plt.show()
```

## MLP Regression

```python
mse = mean_squared_error(y, y_pred_mlp)
r2 = r2_score(y, y_pred_mlp)
print('MLP Regression:')
print(f'  MSE: {mse:.4f}')
print(f'  R² Score: {r2:.4f}')

plt.figure(figsize=(6, 4))
plt.scatter(X, y, color='deepskyblue', label='Actual Data', s=25)
plt.plot(X, y_pred_mlp, color='darkred', label='MLP Predictions', linewidth=2)
plt.plot(X, np.sin(X), color='black', linestyle='--', label='True Sine', linewidth=1)
plt.title('MLP Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('mlp_regression.png')
plt.show()
```
